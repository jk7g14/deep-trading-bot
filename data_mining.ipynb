```python
# Initialising the RNN
ratio = 0.2
model = Sequential()
#model.add(Conv1D(128, kernel_size=5, batch_input_shape=(None,prior,12), activation='relu', padding='valid', strides=1))
# Adding the input layer and the LSTM layer

model.add(CuDNNGRU(batch_input_shape=(None,prior,7), units = 100, return_sequences=False))
model.add(Dropout(ratio))
model.add(BatchNormalization())

#model.add(CuDNNGRU(units = 128, return_sequences=False))
#model.add(Dropout(ratio))
#model.add(BatchNormalization())

#model.add(CuDNNLSTM(units = 128, return_sequences=False))
#model.add(Dropout(0.25))
#model.add(LSTM(batch_input_shape=(None,50,1), units = 4, activation = 'linear',return_sequences=False))
model.add(RepeatVector(after))
#model.add(LSTM(units = 4, activation = 'relu',return_sequences=True))
#model.add(CuDNNGRU(units = 128, return_sequences=True))
#model.add(Dropout(ratio))
#model.add(BatchNormalization())

model.add(CuDNNGRU(units = 200, return_sequences=True))
model.add(Dropout(ratio))
model.add(BatchNormalization())

#model.add(Dropout(0.25))
#model.add(Dense(100, activation = 'linear'))
#model.add(Dense(30, activation = 'linear'))
# Adding the output layer
#model.add(TimeDistributed(Dense(200)))
#model.add(TimeDistributed(Dense(100)))

model.add(TimeDistributed(Dense(2, activation = 'softmax')))

# Compiling the RNN
rmsprop = RMSprop(lr=0.001, rho=0.9, epsilon=1e-8, decay=1e-7)
#adam = Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=1e-8, decay=1e-6, amsgrad=False)
model.compile(optimizer = rmsprop, loss = 'categorical_crossentropy', metrics=['accuracy'])
tensorboard = TensorBoard(log_dir='logs/test6')
model.summary()
```

```python
# Fitting the RNN to the Training set
model.fit(train_x, to_categorical(train_y,2), batch_size = 25, epochs = 300, validation_split=0.1, shuffle=True,callbacks=[tensorboard])
```
